```python
import pandas as pd
import seaborn as sb
```

```python
raw = pd.read_csv("mukamel.txt", sep='\t')
```

```python
def get_coords(x):
    data = x.split('_')
    chrom, start, end = data[2:]
    return chrom, start, end
coords = raw['VNTR (hg38)'].apply(get_coords)
```

```python
data = pd.concat([pd.DataFrame(list(coords), columns=['chrom', 'start', 'end']), raw], axis=1)
```

```python
data
```

```python
data[['chrom', 'start', 'end', 'Gene', "Repeat unit", "Exon overlap pattern"]].to_csv("muk.bed", sep='\t', header=False, index=False)
```

```python
patho = pd.read_csv("Patho.tsv", sep='\t')
```

```python
muka = data.rename(columns={'Gene': 'Locus', 'Repeat unit': "Motifs 1", "Exon overlap pattern":"Region"})
muka["Motifs 2"] = None
muka["Path. repeats"] = None
muka["Disease"] = None
muka["Inheritance mode"] = None
muka["Repeat type"] = "VNTR"
```

```python
full = pd.concat([patho, muka[patho.columns]])
full
```
