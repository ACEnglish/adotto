```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
small_data = pd.read_csv("select_small.txt", sep='\t')
small_data['rank'] = small_data.index
small_data["size regime"] = "<10bp"
big_data = pd.read_csv("select_big.txt", sep='\t')
big_data['rank'] = big_data.index
big_data["size regime"] = ">=10bp"
data = pd.concat([small_data, big_data]).reset_index()
```

```python
plt.figure(dpi=180)
p = sb.lineplot(data=data, x="rank", y="pct_captured", hue="size regime")
plt.suptitle("Percent of variants captured from Utmos greedySelect")
plt.title("86 sample long-read haplotype resolve assemblies", fontsize=10)
p.set(xlabel="samples")
plt.grid()
plt.savefig("UtmosPlot.png", facecolor='white', transparent=False)
```
